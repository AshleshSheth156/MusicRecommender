```python
import pandas as pd
import numpy as np
```

## Read the data

```python
songs = pd.read_csv('song_data.csv')
```

```python
songs.head()
```

```python
triplets = pd.read_csv('triplets_file.csv')
```

```python
triplets.head()
```

```python
song_df = pd.merge(songs, triplets.drop_duplicates(['song_id']),how = 'left', on='song_id')
song_df
```

## Check Null and Duplicate Values

```python
print(songs.isnull().sum())
```

```python
songs.duplicated().sum()
```

## Create a column combining title and artist

```python
song_df['song'] = song_df['title']+' - '+song_df['artist_name']
song_df.head()
```

## Collaborative Filtering

```python
song=song_df.drop_duplicates(['user_id','song_id','title','artist_name'])
song=song.reset_index(drop=True)
```

```python
song_pivot=song.pivot(index='user_id',columns='song_id',values='listen_count')
```

```python
# Step 1: Remove duplicates for 'user_id'
song_unique_users = song.drop_duplicates(subset='user_id')

# Step 2: Pivot the DataFrame
song_pivot = song_unique_users.pivot(index='user_id', columns='song_id', values='listen_count')
```

```python
song_pivot=song_pivot.fillna(0)
```

```python
song_pivot.head()
```

# Train a Knn Model

```python
from sklearn.neighbors import NearestNeighbors
knn=NearestNeighbors(n_neighbors=10,metric='cosine')
Model=knn.fit(song_pivot)
```

```python
song_pivot[song_pivot.index == 1]
```

## Predict Neighbors


```python
user = song_pivot.iloc[1,]
```

```python
distances, indices = Model.kneighbors([user])
distances, indices
```

```python
neighbors=[]
for item in indices[0][1:]:
    neighbors.append(song[song.index == item].user_id.values[0])
neighbors
```

```python
neighbor_songs=pd.DataFrame(columns=['user_id','song_id','listen_count','title','artist','song'])
```

```python
neighbor_songs_list = []
for item in neighbors:
    neighbor_songs_list.append(song[song['user_id'] == item])

neighbor_songs = pd.concat(neighbor_songs_list, ignore_index=True)
neighbor_songs
```

```python
neighbor_songs=pd.DataFrame({'Count':neighbor_songs['listen_count'],'Song':neighbor_songs['song']})
```

```python
neighbor_songs=neighbor_songs.sort_values('Count',ascending=False)
neighbor_songs
```

```python
neighbor_songs=neighbor_songs.reset_index(drop=True)
neighbor_songs
```

```python
Recommended_Song=neighbor_songs['Song'][0:10]
print(Recommended_Song)
```
